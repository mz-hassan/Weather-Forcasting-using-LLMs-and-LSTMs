# Air Quality Prediction using LLMs

```python
import re
import json
import requests
import pandas as pd
```

## Functions to generate response using Ollama's REST API

```python
def generate_response(model, question, format, seed=None):
    url = "http://localhost:11434/api/generate"
    
    if seed is not None:
        data = {
            "model": model,
            "prompt": question,
            "format": format,
            "options": {
                "seed": seed,
                "temperature": 0
              }
        }
    else:
        data = {
            "model": model,
            "prompt": question,
            "format": format,
        }
        
    
    response = requests.post(url, json=data)
    
    res_str = ''
    for r in response:
        res_str += r.decode('utf8')
    
    final_str = ''
    for s in res_str.split('\n'):
        if len(s) > 0:
            final_str += json.loads(s)['response'] 
    
    return final_str.strip()

question = 'What color is the sky at different times of the day? Respond using JSON'
print(generate_response('llama2', question, 'json'))
```

## Import CSV file (Delhi ITO)

```python
df = pd.read_csv('Raw_data_1Day_2022_site_117_ITO_Delhi_CPCB_1Day.csv', parse_dates=['Timestamp'])
df.head()
```

## Taking only the first 5 parameters due to context length limitations (and many missing values in others)

```python
df = df[df.columns[:6]]
columns = [col.split()[0] for col in df.columns[:6]]
df.columns = columns
df.head()
```

## Define input prompt

```python
input_prompt = f"""
You are an air quality expert. You will be given air quality data from the ITO station in Delhi. The task is to predict the next day's parameters given the present and previous day conditions. 

For example, if the input is:
{str(df.iloc[0:10]).strip()}

Then, the next 10 day's prediction will be:
{str(df.iloc[10:20].to_json()).strip()}

Find some kind of relation in the input values above to predict the next 10 day's parameters. If you cannot find a relation, make a guess.
{str(df.iloc[-110:-10]).strip()}

Respond using JSON
"""
print(input_prompt)
```

## Model inference 

### Setting seed for reproducibility (sometimes formatting is not right)

```python
llama2_pred = generate_response('llama2', input_prompt, format='json', seed=41)
llama2_pred = json.loads(llama2_pred)
llama2_pred = pd.DataFrame(llama2_pred)
llama2_pred
```

```python
phi_pred = generate_response('phi', input_prompt, format='json', seed=100)
phi_pred = json.loads(phi_pred)
phi_pred = pd.DataFrame(phi_pred)
phi_pred
```

### Mistral has ongoing issue with json format output constraint (hangs indefinitely). [See here](https://github.com/ollama/ollama/issues/1910)
### Using regular format and then post-processing

```python
mistral_pred = generate_response('mistral', input_prompt, format=None, seed=120)
print(mistral_pred)
```

```python
start = mistral_pred.find('{')
end = len(mistral_pred) - mistral_pred[::-1].find('}')
mistral_pred = mistral_pred[start-1:end]
mistral_pred = json.loads(mistral_pred.strip())
mistral_pred = pd.DataFrame(mistral_pred)
mistral_pred
```

## Evaluate metrics

```python
from sklearn.metrics import mean_absolute_error

# ground truth values
gt = df.iloc[-10:]
gt
```

```python
cols = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx']
def calc_mae(gt, pred):
    for col in cols:
        print(f"{col}: {mean_absolute_error(gt[col], pred[col])}")
    print(f"Overall: {mean_absolute_error(gt[cols], pred[cols])}")    
```

```python
calc_mae(gt, llama2_pred)
```

```python
calc_mae(gt, phi_pred)
```

```python
calc_mae(gt, mistral_pred)
```

## Visualize Results

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error

def get_mae(gt, pred):
    mae_values = {}
    
    for col in cols:
        mae = mean_absolute_error(gt[col], pred[col])
        mae_values[col] = mae

    overall_mae = mean_absolute_error(gt[cols], pred[cols])
    mae_values['Overall'] = overall_mae
    
    return mae_values

def plot_mae(df, ax):

    # Calculate MAE for each set of predictions
    mae_values_pred1 = get_mae(df, llama2_pred)
    mae_values_pred2 = get_mae(df, phi_pred)
    mae_values_pred3 = get_mae(df, mistral_pred)

    # Create a DataFrame to hold the MAE values for each prediction
    mae_df = pd.DataFrame({
        'Llama2': mae_values_pred1.values(),
        'Phi': mae_values_pred2.values(),
        'Mistral': mae_values_pred3.values()
    }, index=mae_values_pred1.keys())

    # Add a row for overall MAE
    mae_df.loc['Overall'] = [mae_values_pred1['Overall'], mae_values_pred2['Overall'], mae_values_pred3['Overall']]

    # Plotting the multi-bar graph
    ax = mae_df.plot(kind='bar', figsize=(12, 8))
    ax.set_title('Mean Absolute Error (MAE) Comparison')
    ax.set_ylabel('MAE')
    ax.set_xlabel('Metrics')

    plt.show()
```

```python

import matplotlib
import matplotlib.pyplot as plt
```

```python
def plot_comparison(ax, df, col):
    ax.plot(df['Timestamp'][-30:], df[col][-30:], label='Ground truth')
    ax.plot(df['Timestamp'][-10:], llama2_pred[col], label='LLaMa 2')
    ax.plot(df['Timestamp'][-10:], phi_pred[col], label='Phi')
    ax.plot(df['Timestamp'][-10:], mistral_pred[col], label='Mistral')
    ax.set_title(f'Comparative plots for predicted {col} trends (Zoomed in for clarity)')
    ax.legend()
    ax.set_xlabel('Timestamp')
    ax.set_ylabel(col)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(30, 18))

# Flatten the 2D array of axes into a 1D array
axes = axes.flatten()

for i, col in enumerate(cols):
    plot_comparison(axes[i], df, col)

plot_mae(df[-10:], axes[5])

plt.tight_layout()
plt.show()
```
